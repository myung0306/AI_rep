Function & Distance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5, 5, 101)
x, len(x)
```

```python
x_p=x[x>0]
x_p
```

```python
#1차 함수
a=2
b=3
y=x*a+b

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#2차 함수
a=2
b=1
c=-3
y=(x-b)**2*a+c

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#지수 함수
a=np.e
y=a**x

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#로그 함수
y=np.log(x_p)

plt.plot(x_p, y)
plt.grid()
plt.show()
```

```python
#삼각 함수
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos")
plt.grid()
plt.legend()
plt.show()
```

```python
#tan 함수
y3=np.tan(x)

plt.plot(x, y3, label="tan")
plt.grid()
plt.legend()
plt.show()
```

```python
#Euclidean distance
x=np.array([1, 1])
y=np.array([4, 5])

z=np.stack([x, y])
plt.scatter(z[:, 0], z[:, 1])
plt.grid()
plt.show()

d=np.sqrt(np.sum((x-y)**2))
print(f"distance:{d}")
```

```python
#Manhattan distance
x=np.array([1, 1])
y=np.array([4, 5])

z=np.array([x, y])
plt.scatter(z[:, 0], z[:, 1])
plt.grid()
plt.show()

d=np.sum(np.abs(x-y))
print(f"distance: {d}")
```

```python
x=np.linspace(0, 1, 1001)
y=-np.log(x)

plt.plot(x, y, label="-ln(x)")
plt.grid()
plt.legend()
plt.show()
```

Vector: 크기와 방향 모두 가지고 있음
Vector: 세기+방향
세기와 방향이 같으면 같은 벡터

```python
#Vector [3, 4]의 스칼라 연산
x1=np.array([3, 4])
x2=2*x1
x3=0.5*x1
x4=-1*x1
x5=-2*x1
x6=-0.5*x1

z=np.stack([x1, x2, x3, x4, x5, x6])
plt.scatter(z[:, 0], z[:, 1])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
#Vector u=[4, 2], v=[1, 3]의 더하기 연산
u=np.array([4, 2])
v=np.array([1, 3])
w=u+v

z=np.stack([u, v, w])
plt.scatter(z[:, 0], z[:, 1])
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.show()
```

```python
#Vector u=[4, 3] L1, L2 Norm
u=np.array([4, 3])

print("L1 norm: ", np.sum(np.abs(u)))
print("L1 norm: ", np.linalg.norm(u, 1))
print("L2 norm: ", np.sqrt(np.sum(u**2)))
print("L2 norm: ", np.linalg.norm(u, 2))
```

```python
#Unit Vector
#Vector의 L2 Norm이 1인 Vector

u=np.array([4, 3])
u_hat=u/np.linalg.norm(u, 2)
print(u)
print("L2 norm:", np.linalg.norm(u, 2))
print(u_hat)
print("L2 norm:", np.linalg.norm(u_hat, 2))
```

```python
#Vector [4, 3], [1, 3]의 Dot Product 및 cos값
u=np.array([4, 3])
v=np.array([1, 3])

print(np.sum(u*v))
print(np.dot(u, v))
print(np.dot(u, v)/(np.linalg.norm(u, 2)*np.linalg.norm(v, 2)))
```

```python
u=np.array([5, 7])
v=np.array([-1, 4])

print("u+v=", u+v)
print("u-v=", u-v)
print("L1 norm:", np.linalg.norm(u, 1))
print("L1 norm:", np.sum(np.abs(u)))
print("L2 norm:", np.linalg.norm(u, 2))
print("L2 norm:", np.sqrt(np.sum(u**2)))

print("L1 norm:", np.linalg.norm(v, 1))
print("L2 norm:", np.linalg.norm(v, 2))

print("u*v=", np.dot(u, v))
print("cos T=", np.dot(u, v)/(np.linalg.norm(u, 2))*(np.linalg.norm(v, 2)))
```
